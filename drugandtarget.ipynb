```python
import pandas as pd
```

```python
df1 = pd.read_csv('all.csv')
df2 = pd.read_csv('drug_sequences.csv')
```

```python
df1.head()
```

```python
# hapus kata drugbank_drug| pada kolom drugbank_id
df2['id'] = df2['id'].str.replace('drugbank_drug|', '')
df2['id'] = df2['id'].str.replace('|', '')
```

```python
df2.head()
```

```python
df2['target'] = ""
for i in range(len(df2)):
    for j in range(len(df1)):
        if df2['id'][i] in df1['Drug IDs'][j]:
            df2['target'][i] = df1['Name'][j]
```

```python
df2.head()
```

```python
df2.to_csv('drug_sequences_target.csv', index=False)
```
